```python
import pandas as pd
import glob
import os
```

```python
# Path to your folder containing CSV files
folder_path = 'data_10/'
```

```python
# Get all CSV file paths in the folder
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
print(f"Found {len(csv_files)} CSV files.")
```

```python
# Read all CSVs and combine them into one big DataFrame
df_list = [pd.read_csv(file) for file in csv_files]
merged_df = pd.concat(df_list, ignore_index=True)
```

```python
merged_df.shape
```

```python
mixed_df = merged_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
mixed_df.shape
```

```python
mixed_df.drop_duplicates(inplace=True)
mixed_df.dropna(inplace=True)
```

```python
mixed_df.to_csv('merged_shuffled_data_10.csv', index=False)
```
